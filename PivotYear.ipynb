```python
import pandas as pd
import numpy as np
import winsound

# Importing data
df = pd.read_csv("PghRestaurantAllYear.csv",dtype=str)
df
```

```python
# Create field
df['Y2019'] = np.where(df['Version Year'] == '2019', 1, 0)
df['Y2020'] = np.where(df['Version Year'] == '2020', 1, 0)
df['Y2021'] = np.where(df['Version Year'] == '2021', 1, 0)
df
```

```python
df = df.groupby(["IUSA Number"]).sum().reset_index()
df
```

```python
#export
df.to_csv("IUSAYear" + ".csv",index=False)
```

```python
df_rest = pd.read_csv("PghRestaurantAllYear.csv",dtype=str)
df_rest = df_rest.drop(columns=['Version Year'])
df_rest = df_rest.drop_duplicates(subset='IUSA Number', keep='last')
#Join
df_join = pd.merge(df, df_rest, on='IUSA Number', how='left')
df_join
```
